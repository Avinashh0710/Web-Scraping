```python
import requests
from bs4 import BeautifulSoup
import openpyxl
```

```python
from google.colab import files
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet.title = "Sharda"
sheet.append(["Title", "Link"])
url = "https://www.sharda.ac.in/"
response = requests.get(url)
if response.status_code !=200:
    print("page not Found",response.status_code)
else:
   soup= BeautifulSoup(response.text,"html.parser")
   links= soup.find_all("a")
   for a in links:
    Title = a.get_text().strip()
    href = a.get("href")
    if Title and len(Title)>20 and href:
        if href.startswith("/"):
            href = url+href
        sheet.append([Title,href])
filename = "webscrapping.xlsx"
workbook.save(filename)
print(f"all the data saved in file", filename)
files.download(filename)
```
